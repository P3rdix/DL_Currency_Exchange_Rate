```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data.csv")
```

```python
data.drop("Unnamed: 0.1", inplace=True, axis=1)
```

```python
data.columns
```

```python
data.tail(1)
```

```python
data.columns = ['Date', 'Australia', 'Europe', 'Brazil', 'Canada',
       'China', 'Denmark', 'Hong Kong', 'India', 'Japan', 'Malaysia']
```

```python
idx = pd.date_range("2010-01-04", "2019-12-31")
data["Date"] = pd.to_datetime(data["Date"])
```

```python
data.info()
```

```python
data.set_index("Date", inplace=True)
```

```python
data = data.reindex(idx, fill_value=0)
```

```python
dt = data[data!=0]
```

```python
fig, axs = plt.subplots(1,1, figsize=(18,5))
exclude = ["Japan","India", "Hong Kong"]
x = data.index
for i in data.columns:
    if i not in exclude:
        axs.plot(x,dt[i],label=i)
axs.legend()
fig.show()
```
